```python
import os
from getpass import getpass
from langchain import HuggingFaceHub
```

```python
HUGGINGFACEHUB_API_TOKEN = getpass()
```

```python
os.environ['HUGGINGFACEHUB_API_TOKEN'] = HUGGINGFACEHUB_API_TOKEN
```

```python
llm = HuggingFaceHub(repo_id = 'mistralai/Mixtral-8x7B-Instruct-v0.1')
llm
```

```python
from langchain.retrievers.multi_query import MultiQueryRetriever
from langchain_community.document_loaders import WebBaseLoader
from langchain.vectorstores import Chroma
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = WebBaseLoader("https://lilianweng.github.io/posts/2023-06-23-agent/")
data = loader.load()
data
```

```python
text_splitters = RecursiveCharacterTextSplitter(chunk_size = 500,chunk_overlap = 0)
splits = text_splitters.split_documents(data)
splits
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embedding = HuggingFaceEmbeddings()
vectorbdb = Chroma.from_documents(documents = splits , embedding = embedding)
```

```python
Question = "What are the approches to task decomposition "
retriever_from_llm = MultiQueryRetriever.from_llm(retriever = vectorbdb.as_retriever(), llm =llm)
```

```python
import logging
logging.basicConfig()
logging.getLogger("langchain.retrievers.multi_query").setLevel(logging.INFO)
```

```python
unique_docs = retriever_from_llm.get_relevant_documents(query = Question)
```
