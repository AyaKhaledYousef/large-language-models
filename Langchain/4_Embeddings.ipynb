# Store Embedding in ChromaDB

```python
# import
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings.sentence_transformer import (
    SentenceTransformerEmbeddings,
)
from langchain_community.vectorstores import Chroma
from langchain_text_splitters import CharacterTextSplitter
```

```python
# load the document and split it into chunks
loader = TextLoader("/content/Sentences_AllAgree.txt", encoding='latin-1')
documents = loader.load()
documents[0].page_content
```

```python

# split it into chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)

len(docs)
```

```python
# create the open-source embedding function
embedding_function = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
embedding_function
```

```python
# load it into Chroma
db = Chroma.from_documents(docs, embedding_function)

# query it
query = "cargo revenue increased by 5.4 % to EUR 21.46 mn , and cargo volume increased by 2.4 % to 70,116 cargo units "
result = db.similarity_search(query,k=1)

print (len(result))
print(docs[0].page_content)
```
